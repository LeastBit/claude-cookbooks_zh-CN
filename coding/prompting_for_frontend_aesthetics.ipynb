# 前端美学：提示工程指南

Claude 可以生成高质量的前端，但缺乏指导时往往会倾向于通用、保守的设计。本指南将展示如何提示 Claude 产生更独特、精致的输出。

## 提示更好的输出

Claude 具备设计原则、排版和色彩理论的强大知识，但默认会选择安全的选择，除非明确鼓励其他做法。通过实验，我们发现了三个能持续产生更好结果的策略：

1. **引导特定的设计维度** - 分别引导 Claude 的注意力到排版、颜色、动效和背景
2. **参考设计灵感** - 建议像 IDE 主题或文化美学这样的来源，但不要过于规定性  
3. **点名常见的默认值** - 明确告诉 Claude 避免其趋向于通用选择的倾向

下面的提示将这三个策略应用于四个关键设计领域。

## 提示词

要实现这些更改，您可以将此提示部分附加到您的系统提示或 CLAUDE.md 文件中。

```python
DISTILLED_AESTHETICS_PROMPT = """
<frontend_aesthetics>
你倾向于收敛到通用的、"分布上"的输出。在前端设计中，这会产生用户所谓的"AI 垃圾"美学。避免这种情况：制作有创意、独特的前端，给人惊喜和愉悦。专注于：

排版：选择美观、独特、有趣的字体。避免使用 Arial 和 Inter 等通用字体；选择能提升前端美学的独特选择。

颜色和主题：坚持一致的美学风格。使用 CSS 变量保持一致性。带有锐利点缀的主要色彩优于胆怯、均匀分布的调色板。从 IDE 主题和文化美学中汲取灵感。

动效：使用动画来营造效果和微交互。HTML 优先使用纯 CSS 解决方案。如可用，在 React 中使用 Motion 库。专注于高影响时刻：一次精心策划的页面加载，带有交错显示（animation-delay），比分散的微交互创造更多愉悦。

背景：创造氛围和深度，而不是默认使用纯色。叠加 CSS 渐变，使用几何图案，或添加与整体美学匹配的上下文效果。

避免通用 AI 生成的审美：
- 过度使用的字体系列（Inter、Roboto、Arial、系统字体）
- 陈词滥调的配色方案（特别是白色背景上的紫色渐变）
- 可预测的布局和组件模式
- 缺乏上下文特定特色的千篇一律设计

创造性地解释并做出感觉真正为上下文设计的意外选择。在浅色和深色主题、不同字体、不同美学之间变化。你仍然倾向于在代际之间收敛到常见选择（例如 Space Grotesk）。避免这种情况：跳出框框思考至关重要！
</frontend_aesthetics>
"""
```
